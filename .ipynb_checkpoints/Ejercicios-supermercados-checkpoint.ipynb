- Ejercicio 4

```python
import re


data["mes"] = pd.to_datetime(data["fecha"], format = "%d/%m/%Y") + pd.offsets.MonthEnd(0)

ejercicio4 = pd.pivot_table(data, index = ["mes"], columns = ["SECTOR"], values = ["precio", "importe"], aggfunc= {"importe": np.sum, "precio": np.mean})

ejercicio4.columns = ejercicio4.columns.droplevel()
columnas = ejercicio4.columns
regex = re.compile("\d\s")

columnas = [e if len(regex.findall(e)) == 0 else e[2:] for e in columnas]

parteFinal = ["_sum_importe"]*4 + ["_prom_precio"]*4


nuevasColumnas = [e1 + e2 for e1, e2 in zip(columnas, parteFinal)]

ejercicio4.columns = nuevasColumnas
ejercicio4 = ejercicio4[['COMESTIBLES_prom_precio', 'COMESTIBLES_sum_importe', 'BEBIDAS_prom_precio','BEBIDAS_sum_importe',
            'NO COMESTIBLES_prom_precio', 'NO COMESTIBLES_sum_importe', 
            'PERECEDEROS_prom_precio', 'PERECEDEROS_sum_importe']]

ejercicio4 = ejercicio4.sort_index()

ejercicio4.index = pd.Series(ejercicio4.index).dt.month_name(locale = "Spanish")

ejercicio4
```
